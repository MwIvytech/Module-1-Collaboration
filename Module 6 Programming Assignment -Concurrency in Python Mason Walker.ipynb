```python
# 13.1
from datetime import date

today = date.today()

today_str = str(today)

with open("today.txt", "w") as file:
    file.write(today_str)
```

```python
# 13.2
with open("today.txt", "r") as file:
    today_string = file.read()

print(today_string)
```

```python
from datetime import datetime
today_date = datetime.strptime(today_string,  "%Y-%m-%d").date()

print(today_date)
```

```python
# 15.1
import multiprocessing
import time
import random
from datetime import datetime

def generator(name):
    delay = random.uniform(0, 1)
    time.sleep(delay)
    print(f"Process {name} finished at {datetime.now().strftime('%H:%M:%S.%f')} after waiting {delay:.3f} seconds")

if __name__ == "__main__":
    processes = []
    for i in range(3):
        p = multiprocessing.Process(target=generator, args=(i+1,))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()

    print("Complete")
```
