```python
# 7.5 Capitalize an element , Yes this changed the element in the list
things = ["mozzarella", "cinderella", "salmonella"]
things[1] = things[1].capitalize()
print(things)
```

```python
# 7.6 Make an element all uppercase
things = ["mozzarella", "cinderella", "salmonella"]
things[0] = things[0].upper()
print(things)
```

```python
# 7.7 delete an element 
things = ["mozzarella", "cinderella", "salmonella"]
del things[2]
print(things)
```

```python
# 9.2 define generator for odd numbers
def get_odds():
    for number in range(10):
        if number % 2 == 1:
            yield number

# use loop to find and print the 3rd value found 
count = 0
for odd in get_odds():
    count+=1
    if count == 3:
        print(odd)
        break
```
