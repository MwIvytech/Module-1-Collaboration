```python
# Sort an Array of 0s, 1s, and 2s
def sorter(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low +=1
            mid +=1
        elif arr[mid] ==1:
            mid +=1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -=1
    return arr

print(sorter([0, 0, 2, 1, 2, 0, 1, 2, 1, 1, 0, 2]))
```

```python
# Sort with binary search
def binarysearch(arr, k):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == k:
            result = mid
            high = mid -1
        elif arr[mid] < k:
            low = mid + 1
        else:
            high = mid - 1

    return result

print(binarysearch([1, 2, 3, 4, 5], 4))
print(binarysearch([11, 22, 33, 44, 55], 445))
print(binarysearch([1, 1, 1, 1, 2], 1))
```
