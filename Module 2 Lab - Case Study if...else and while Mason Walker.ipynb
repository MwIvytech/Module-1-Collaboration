```python
# Author: Mason Walker
# File name: Module 2 Lab - Case Study: if...else and while
# Description: This program accepts student names and GPA, then checks if any honors qualify

while True:
    last_name = input("Please enter student's last name. Press 'ZZZ' to quit.")
    if last_name.upper() == 'zzz':
        print("Good Bye")
        break
    first_name = input("Please enter student's first name.")
    gpa = float(input("Please enter student's gps."))
    if gpa <= 0 or gpa > 4:
        print("invalid GPA.")
        continue

    if gpa >= 3.5:
        print("Dean's List.")
    elif gpa >= 3.25:
        print("Honor Roll.")
    else:
        print("No Honors this time.")
    
```
